```python
my_file = open("data/Market_Basket_Optimisation.csv", "r")
lines = my_file.readlines()
lines = [line.replace('\n','') for line in lines]
dataset = [line.split(',') for line in lines]
dataset[:5]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
oht_ary = te.fit(dataset).transform(dataset, sparse=True)
df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
df
```

```python
#http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.02,use_colnames=True)
frequent_itemsets
```

```python
#http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
rules.iloc[:,[0,1,4,5,6]]
```

```python
rules_2 = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_2.iloc[:,[0,1,4,5,6]]
```
