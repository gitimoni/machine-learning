```python
items_list = []
for indx in range(0, df.shape[0], 1):
    items_list.append([x for x in df.iloc[indx] if str(x) != 'nan'])
items_list[:10]
```

```python
#importing the model
from apriori_lib import apriori
rules = apriori(items_list, min_support = 0.01, min_confidence = 0.3, min_lift = 1.5)
result = list(rules)
print(len(result))
print(result)
```

```python
transactions = [['i1', 'i2', 'i3', 'i5'], 
              ['i2', 'i4'], 
              ['i2', 'i3', 'i4'], 
              ['i1', 'i2', 'i4'],
             ['i1', 'i3', 'i4'],
             ['i1', 'i2', 'i3', 'i4'],
             ['i1', 'i2', 'i3', 'i4'],
             ['i1', 'i2', 'i3', 'i5'],
             ['i1', 'i2', 'i3']]
from apriori_lib import apriori
rules = apriori(transactions, min_support = 0.5, min_confidence = 0.5, min_lift = 1.01)
result = list(rules)
#print(result)
for i in range(0, len(result), 1):
    print('\n', result[i])
```
