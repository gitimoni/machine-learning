```python
import pandas as pd

df = pd.read_csv('data/Data.csv')

df
```

```python
df.fillna(
            {
                'Age': df.Age.median(),
                'Salary': df.Salary.mean()
            }, 
            inplace = True
            )

df
```

```python
df = pd.get_dummies(df, ['Country','Purchased'], drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(df)
```

```python
scaler.transform(df)
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled
#scaler.inverse_transform(df_scaled) for reverse scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_standardised = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
df_standardised
```

```python
df_standardised.Age[df_standardised.Age>3]=df_standardised.Age.median()
df_standardised.Age[df_standardised.Age<-3]=df_standardised.Age.median()
```
